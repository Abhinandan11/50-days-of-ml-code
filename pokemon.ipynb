```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('datasets/Pokemon.csv', index_col = 0)
```

# lmplot

```python
sns.lmplot(x = 'HP', y = 'Total', data = df, hue = 'Stage')
```

```python
sns.lmplot(x = 'HP', y = 'Total', data = df, fit_reg = False, col = 'Stage')
```

```python
sns.lmplot(x = 'Attack', y = 'Total', data = df, fit_reg = False, col = 'Stage')
```
