```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = 11.7,8.27
```

```python
df = pd.read_csv('datasets/Pokemon.csv', index_col = 0)
```

```python
df.head()
```

# Countplot

```python
sns.countplot(x = 'Type 1', data = df)
```

# Distplot

```python
sns.distplot(df.HP)
```

```python
sns.distplot(df.Attack)
```

# lmplot

```python
sns.lmplot(x = 'HP', y = 'Total', data = df, hue = 'Stage')
```

```python
sns.lmplot(x = 'HP', y = 'Total', data = df, fit_reg = False, hue = 'Stage')
```

```python
sns.lmplot(x = 'HP', y = 'Total', data = df, fit_reg = False, col = 'Stage')
```

```python
sns.lmplot(x = 'Attack', y = 'Total', data = df, fit_reg = False, col = 'Stage')
```

```python
sns.lmplot(x = 'Defense', y = 'Total', data = df, fit_reg = False, col = 'Stage')
```

```python
sns.lmplot(x = 'Attack', y = 'Total', data = df, fit_reg = False, hue = 'Type 1')
```

```python
sns.lmplot(x = 'Defense', y = 'Total', data = df, fit_reg = False, hue = 'Type 1')
```

```python
sns.lmplot(x = 'Speed', y = 'Total', data = df, fit_reg = False, hue = 'Type 1')
```

# Boxplot

```python
sns.boxplot(data = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']])
```

# Violinplot

```python
sns.violinplot(x = 'Type 1', y = 'Attack', data = df)
```

```python
sns.violinplot(x = 'Type 1', y = 'Defense', data = df)
```

```python
sns.violinplot(x = 'Type 1', y = 'Speed', data = df)
```

# Swarmplot

```python
sns.swarmplot(x = 'Type 1', y = 'HP', data = df)
```

```python
sns.swarmplot(x = 'Type 1', y = 'HP', data = df, hue = 'Stage')
```

# Heatmap

```python
stats_df = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
```

```python
corr = stats_df.corr()
```

```python
sns.heatmap(corr, cmap = 'Greens')
```
